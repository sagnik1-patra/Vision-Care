```python
import os
import sys
import json
import time
import glob
import traceback
from typing import List, Tuple, Optional, Dict
from dataclasses import dataclass, asdict
from collections import defaultdict

import numpy as np
import pandas as pd
import cv2
import h5py
import yaml
from tqdm import tqdm
from skimage.feature import local_binary_pattern

# --------------------------
# CONFIG: WINDOWS PATHS
# --------------------------
ROOT_DIR = r"C:\Users\NXTWAVE\Downloads\Vision Care\archive\Odir5k preprocessed with CLAHE\training images"
OUT_DIR  = r"C:\Users\NXTWAVE\Downloads\Vision Care"

# Image processing / feature params
IMG_SIZE = (256, 256)          # resize target (w, h); fundus images benefit from slightly larger than 128
LBP_P = 8                      # neighbors
LBP_R = 1                      # radius
LBP_METHOD = "uniform"         # 'uniform' -> bins = P + 2

# File extensions
IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff", ".webp"}

# --------------------------
# Data structures
# --------------------------
@dataclass
class Item:
    id: str
    path: str
    label_name: str
    label_id: int
    width: Optional[int] = None
    height: Optional[int] = None
    extra: Optional[dict] = None

# --------------------------
# Utils
# --------------------------
def ensure_dir(path: str):
    if not os.path.exists(path):
        os.makedirs(path, exist_ok=True)

def list_subdirs(path: str) -> List[str]:
    return [os.path.join(path, d) for d in os.listdir(path) if os.path.isdir(os.path.join(path, d))]

def list_files_recursive(dir_path: str) -> List[str]:
    files = []
    for root, _, fnames in os.walk(dir_path):
        for f in fnames:
            files.append(os.path.join(root, f))
    return files

def is_image(path: str) -> bool:
    return os.path.splitext(path)[1].lower() in IMG_EXTS

def imread_unicode_color(path: str) -> Optional[np.ndarray]:
    # Windows-safe reader that supports unicode/spaces
    data = np.fromfile(path, dtype=np.uint8)
    if data.size == 0:
        return None
    img = cv2.imdecode(data, cv2.IMREAD_COLOR)
    return img

def resize_keep_aspect(img: np.ndarray, target_wh: Tuple[int, int]) -> np.ndarray:
    tw, th = target_wh
    h, w = img.shape[:2]
    scale = min(tw / w, th / h)
    nw, nh = int(w * scale), int(h * scale)
    resized = cv2.resize(img, (nw, nh), interpolation=cv2.INTER_AREA)
    # pad to target
    canvas = np.zeros((th, tw, 3), dtype=img.dtype)
    y0 = (th - nh) // 2
    x0 = (tw - nw) // 2
    canvas[y0:y0+nh, x0:x0+nw] = resized
    return canvas

def lbp_histogram(gray: np.ndarray, P: int, R: int, method: str) -> np.ndarray:
    lbp = local_binary_pattern(gray, P, R, method)
    n_bins = P + 2 if method == "uniform" else int(lbp.max() + 1)
    hist, _ = np.histogram(lbp.ravel(), bins=n_bins, range=(0, n_bins), density=True)
    return hist.astype("float32")

def extract_features_bgr(img_bgr: np.ndarray) -> np.ndarray:
    """
    Features: [LBP on green channel (P+2 bins)] + [RGB mean/std (6 dims)] -> total dims = (LBP_P+2) + 6
    """
    # Resize (letterbox) to keep fundus circular FOV roughly centered
    img = resize_keep_aspect(img_bgr, IMG_SIZE)

    # Color stats (mean/std per channel)
    means = img.mean(axis=(0,1))            # B, G, R
    stds  = img.std(axis=(0,1)) + 1e-8
    color_stats = np.concatenate([means, stds], axis=0).astype("float32")

    # LBP on green channel (often best for retinal lesions)
    g = img[:, :, 1]
    g_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # backup gray
    # Prefer green channel; but if something's off, use gray
    green = g if g is not None else g_gray
    lbp_hist = lbp_histogram(green, LBP_P, LBP_R, LBP_METHOD)

    feat = np.concatenate([lbp_hist, color_stats], axis=0).astype("float32")
    return feat

# --------------------------
# Main
# --------------------------
def main():
    t0 = time.time()
    if not os.path.isdir(ROOT_DIR):
        print(f"[ERROR] Training images folder not found:\n  {ROOT_DIR}")
        sys.exit(1)

    ensure_dir(OUT_DIR)

    # Collect class folders (direct children of ROOT_DIR)
    class_dirs = [d for d in list_subdirs(ROOT_DIR)]
    if not class_dirs:
        print(f"[ERROR] No class subfolders found under:\n  {ROOT_DIR}\n"
              f"Expected layout: training images/<class_name>/*.jpg")
        sys.exit(2)

    # Label mapping in alphabetical order of folder names (stable & reproducible)
    class_dirs_sorted = sorted(class_dirs, key=lambda p: os.path.basename(p).lower())
    label_names = [os.path.basename(p).lower() for p in class_dirs_sorted]
    label_to_id: Dict[str, int] = {name: i for i, name in enumerate(label_names)}
    print("[INFO] Label map:", label_to_id)

    # Walk files
    records: List[Tuple[str, str, int]] = []
    per_class_counts = defaultdict(int)
    for cdir in class_dirs_sorted:
        cname = os.path.basename(cdir).lower()
        files = list_files_recursive(cdir)
        usable = 0
        for f in files:
            if is_image(f):
                records.append((f, cname, label_to_id[cname]))
                usable += 1
        per_class_counts[cname] = usable
        print(f"[CLASS] {cname:18s} -> total files {len(files):5d}, usable images {usable:5d}")

    if not records:
        print("[ERROR] No images found (supported exts:", ", ".join(sorted(IMG_EXTS)), ")")
        sys.exit(3)

    # Extract features
    feats = []
    labels = []
    paths = []
    widths = []
    heights = []
    items: List[Item] = []

    print("[INFO] Extracting features...")
    for idx, (path, lname, lid) in enumerate(tqdm(records)):
        try:
            img = imread_unicode_color(path)
            if img is None:
                continue
            h, w = img.shape[:2]
            feat = extract_features_bgr(img)
            feats.append(feat)
            labels.append(lid)
            paths.append(path)
            widths.append(w)
            heights.append(h)
            items.append(Item(
                id=str(len(paths)-1),
                path=path,
                label_name=lname,
                label_id=lid,
                width=w,
                height=h,
                extra=None
            ))
        except Exception as e:
            print(f"[WARN] Failed on {path}: {e}")

    if len(feats) == 0:
        print("[ERROR] No features extracted. Please verify images are valid.")
        sys.exit(4)

    X = np.stack(feats, axis=0)
    y = np.array(labels, dtype=np.int64)
    print(f"[INFO] Features shape: {X.shape} | Samples: {len(items)} | Classes: {len(label_names)}")

    # --------------------------
    # Save Artifacts
    # --------------------------
    base = "visioncare_index"

    # 1) HDF5
    h5_path = os.path.join(OUT_DIR, f"{base}.h5")
    print(f"[SAVE] HDF5 -> {h5_path}")
    with h5py.File(h5_path, "w") as f:
        f.create_dataset("X", data=X, compression="gzip", compression_opts=4)
        f.create_dataset("y", data=y, compression="gzip", compression_opts=4)
        # filenames
        dt = h5py.special_dtype(vlen=str)
        ds_paths = f.create_dataset("filenames", (len(paths),), dtype=dt)
        ds_paths[:] = paths
        # attrs
        f.attrs["img_size"] = json.dumps({"w": IMG_SIZE[0], "h": IMG_SIZE[1]})
        f.attrs["lbp"] = json.dumps({"P": LBP_P, "R": LBP_R, "method": LBP_METHOD})
        f.attrs["label_map"] = json.dumps({str(i): name for name, i in label_to_id.items()})
        f.attrs["feature_schema"] = json.dumps({
            "order": ["lbp_green_hist", "rgb_means", "rgb_stds"],
            "dims": {"lbp_green_hist": (LBP_P + 2), "rgb_means": 3, "rgb_stds": 3},
            "total_dim": int(X.shape[1])
        })

    # 2) PKL (DataFrame)
    pkl_path = os.path.join(OUT_DIR, f"{base}.pkl")
    print(f"[SAVE] PKL  -> {pkl_path}")
    df = pd.DataFrame(X)
    df["label_id"] = y
    df["label_name"] = [it.label_name for it in items]
    df["path"] = paths
    df["width"] = widths
    df["height"] = heights
    df.to_pickle(pkl_path)

    # 3) YAML (stats + label map)
    yaml_path = os.path.join(OUT_DIR, f"{base}.yaml")
    print(f"[SAVE] YAML -> {yaml_path}")
    stats = {
        "total_items": int(len(items)),
        "classes": int(len(label_names)),
        "by_label": {name: int(per_class_counts[name]) for name in label_names},
        "img_size": {"width": IMG_SIZE[0], "height": IMG_SIZE[1]},
        "lbp": {"P": LBP_P, "R": LBP_R, "method": LBP_METHOD},
        "feature_schema": {
            "order": ["lbp_green_hist", "rgb_means", "rgb_stds"],
            "dims": {"lbp_green_hist": (LBP_P + 2), "rgb_means": 3, "rgb_stds": 3},
            "total_dim": int(X.shape[1])
        },
        "label_map": {name: int(idx) for name, idx in label_to_id.items()}
    }
    with open(yaml_path, "w", encoding="utf-8") as fy:
        yaml.safe_dump(stats, fy, sort_keys=False, allow_unicode=True)

    # 4) JSON (index)
    json_path = os.path.join(OUT_DIR, f"{base}.json")
    print(f"[SAVE] JSON -> {json_path}")
    payload = {
        "version": "1.0",
        "created_at": time.strftime("%Y-%m-%d %H:%M:%S"),
        "label_map": {name: int(idx) for name, idx in label_to_id.items()},
        "items": [asdict(it) for it in items]
    }
    with open(json_path, "w", encoding="utf-8") as fj:
        json.dump(payload, fj, indent=2, ensure_ascii=False)

    t1 = time.time()
    print("\n[SUMMARY]")
    print("Artifacts created in:", OUT_DIR)
    print(f" - HDF5: {h5_path}")
    print(f" - PKL : {pkl_path}")
    print(f" - YAML: {yaml_path}")
    print(f" - JSON: {json_path}")
    print(f"Items: {len(items)}, X shape: {X.shape}, elapsed: {t1 - t0:.1f}s")


if __name__ == "__main__":
    try:
        main()
    except Exception as e:
        print("[FATAL]", e)
        traceback.print_exc()
        sys.exit(99)
```

```python
# Jupyter cell — VisionCare: Accuracy Curve + Confusion Matrix Heatmap
import os, json, warnings, numpy as np, pandas as pd, matplotlib.pyplot as plt, h5py, yaml
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

warnings.filterwarnings("ignore")

# --------------------------
# CONFIG (edit if your path differs)
# --------------------------
OUT_DIR  = r"C:\Users\NXTWAVE\Downloads\Vision Care"
PKL_PATH = os.path.join(OUT_DIR, "visioncare_index.pkl")
H5_PATH  = os.path.join(OUT_DIR, "visioncare_index.h5")
YAML_PATH = os.path.join(OUT_DIR, "visioncare_index.yaml")

VAL_SIZE = 0.2
RANDOM_STATE = 42
NORMALIZE_CM = False   # set True to show percentages in heatmap

# --------------------------
# Helpers
# --------------------------
def load_label_names():
    names = None
    # Prefer YAML (stored as {name: id})
    if os.path.isfile(YAML_PATH):
        try:
            with open(YAML_PATH, "r", encoding="utf-8") as f:
                yml = yaml.safe_load(f)
            lm = yml.get("label_map", None)
            if isinstance(lm, dict):
                inv = {int(v): k for k, v in lm.items()}
                max_id = max(inv) if inv else -1
                names = [inv.get(i, f"class_{i}") for i in range(max_id+1)]
        except Exception:
            pass
    # Fallback: H5 attrs (stored as {"0":"name", ...})
    if names is None and os.path.isfile(H5_PATH):
        try:
            with h5py.File(H5_PATH, "r") as f:
                lm_attr = f.attrs.get("label_map", None)
                if lm_attr:
                    lm = json.loads(lm_attr)
                    inv = {int(k): v for k, v in lm.items()}
                    max_id = max(inv) if inv else -1
                    names = [inv.get(i, f"class_{i}") for i in range(max_id+1)]
        except Exception:
            pass
    return names

def load_data():
    class_names = load_label_names()
    if os.path.isfile(PKL_PATH):
        df = pd.read_pickle(PKL_PATH)
        meta_cols = {"label_id", "label_name", "path", "width", "height"}
        feat_cols = [c for c in df.columns if c not in meta_cols]
        X = df[feat_cols].to_numpy(dtype=np.float32)
        y = df["label_id"].astype(int).to_numpy()
        # fallback names from PKL if needed
        if class_names is None and "label_name" in df.columns:
            id2name = {}
            for lid, lname in zip(df["label_id"], df["label_name"]):
                id2name.setdefault(int(lid), str(lname))
            max_id = int(df["label_id"].max())
            class_names = [id2name.get(i, f"class_{i}") for i in range(max_id+1)]
        return X, y, class_names
    if os.path.isfile(H5_PATH):
        with h5py.File(H5_PATH, "r") as f:
            X = f["X"][:]
            y = f["y"][:].astype(int)
        return X, y, class_names
    raise FileNotFoundError(f"Artifacts not found at:\n  {PKL_PATH}\n  {H5_PATH}")

def build_model():
    # SVM (RBF) baseline with scaling
    return Pipeline([
        ("scaler", StandardScaler(with_mean=True)),
        ("clf", SVC(C=10.0, gamma="scale", kernel="rbf"))
    ])

def show_and_save_accuracy_curve(fold_accs, out_path):
    plt.figure(figsize=(8,5))
    xs = list(range(1, len(fold_accs)+1))
    plt.plot(xs, fold_accs, marker="o", linewidth=2)
    plt.title("VisionCare — Stratified 5-Fold CV Accuracy")
    plt.xlabel("Fold"); plt.ylabel("Accuracy")
    plt.ylim(0, 1.0)
    plt.grid(True, linestyle="--", alpha=0.4)
    mean_acc = float(np.mean(fold_accs))
    plt.axhline(mean_acc, linestyle="--", linewidth=1.5)
    plt.text(xs[-1]-0.2, mean_acc+0.01, f"Mean={mean_acc:.3f}")
    plt.tight_layout()
    plt.savefig(out_path, dpi=200)
    plt.show()

def show_and_save_confusion_matrix(cm, class_names, out_path, normalize=False):
    if normalize:
        with np.errstate(all='ignore'):
            row_sums = cm.sum(axis=1, keepdims=True)
            cm_plot = np.divide(cm, row_sums, where=row_sums!=0)
        fmt = ".2f"; title = "VisionCare — Confusion Matrix (Normalized)"
    else:
        cm_plot = cm.astype(int)
        fmt = "d"; title = "VisionCare — Confusion Matrix (Counts)"

    n = cm_plot.shape[0]
    labels = class_names if class_names is not None else [f"class_{i}" for i in range(n)]

    fig, ax = plt.subplots(figsize=(max(6, n), max(5, n)))
    im = ax.imshow(cm_plot, interpolation="nearest")
    ax.figure.colorbar(im, ax=ax)
    ax.set(xticks=np.arange(n), yticks=np.arange(n),
           xticklabels=labels, yticklabels=labels,
           ylabel="True label", xlabel="Predicted label", title=title)
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

    vmax = np.nanmax(cm_plot) if np.isfinite(cm_plot).any() else 0
    thresh = vmax / 2.0 if vmax > 0 else 0.5
    for i in range(n):
        for j in range(n):
            val = cm_plot[i, j]
            txt = f"{val:{fmt}}" if np.isfinite(val) else "0"
            ax.text(j, i, txt, ha="center", va="center",
                    color="white" if (np.isfinite(val) and val > thresh) else "black")

    fig.tight_layout()
    plt.savefig(out_path, dpi=200, bbox_inches="tight")
    plt.show()

# --------------------------
# Run
# --------------------------
os.makedirs(OUT_DIR, exist_ok=True)

print("[INFO] Loading VisionCare artifacts...")
X, y, class_names = load_data()
print(f"[INFO] X={X.shape}, classes={len(np.unique(y))}, names={class_names}")

# ----- 5-fold CV for accuracy curve -----
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
fold_accs, rows = [], []
for i, (tr, te) in enumerate(skf.split(X, y), start=1):
    model = build_model()
    model.fit(X[tr], y[tr])
    pred = model.predict(X[te])
    acc = accuracy_score(y[te], pred)
    fold_accs.append(acc); rows.append({"fold": i, "accuracy": acc})
    print(f"  Fold {i}: acc={acc:.4f}")

# Save & show accuracy plot
acc_png = os.path.join(OUT_DIR, "visioncare_accuracy_curve.png")
pd.DataFrame(rows).to_csv(os.path.join(OUT_DIR, "visioncare_cv_scores.csv"), index=False)
show_and_save_accuracy_curve(fold_accs, acc_png)
print(f"[SAVE] Accuracy graph -> {acc_png}")

# ----- Held-out split for confusion matrix -----
Xtr, Xva, ytr, yva = train_test_split(X, y, test_size=VAL_SIZE, random_state=RANDOM_STATE, stratify=y)
model = build_model()
model.fit(Xtr, ytr)
yhat = model.predict(Xva)

labels_range = np.arange(len(class_names)) if class_names is not None else np.arange(np.max(y)+1)
cm = confusion_matrix(yva, yhat, labels=labels_range)
cm_png = os.path.join(OUT_DIR, "visioncare_confusion_matrix_heatmap.png")
if class_names is None:
    class_names = [f"class_{i}" for i in range(cm.shape[0])]
show_and_save_confusion_matrix(cm, class_names, cm_png, normalize=NORMALIZE_CM)
print(f"[SAVE] Confusion matrix heatmap -> {cm_png}")

# Classification report
report_txt = os.path.join(OUT_DIR, "visioncare_classification_report.txt")
with open(report_txt, "w", encoding="utf-8") as f:
    f.write(classification_report(yva, yhat, target_names=class_names, digits=4))
print(f"[SAVE] Classification report -> {report_txt}")

print("\n[SUMMARY]")
print(f"  CV mean acc: {np.mean(fold_accs):.4f} (std {np.std(fold_accs):.4f})")
```
